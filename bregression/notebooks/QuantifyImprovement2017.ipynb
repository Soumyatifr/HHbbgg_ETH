```python
whats=['p_T','eta','rho']
which = 0
```

```python
where=''
files = 'data_IQR_compare_%s_%s%s.csv'%(whats[which],sample,where)
df = pd.read_csv(path+files)
```

```python
to_compare=['p_T','delta_Hbb_JEC_rel','delta_ETH_JEC_rel','delta_ETH_JEC_rel','delta_Pisa_JEC_rel','delta_ETH_Pisa_JEC_rel']
#df[to_compare].round(3)
```

```python
ax = df.plot.scatter(x='%s'%whats[which],y='delta_final2016_JEC_rel',color='g',s=91,label='NN 2016 vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_withLepton_JEC_rel',ax=ax,color='b',s=91,label='with lepton vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_withLeptonwoJEC_JEC_rel',ax=ax,color='r',s=91,label='with lepton minus JEC vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_us_pisa_minusLepton_JEC_rel',ax=ax,color='brown',s=91,label='us pisa set3 vs JEC')


ax.set_ylabel('$\\frac{\Delta \sigma/\mu}{\sigma/\mu_{JEC}}$ ',fontsize=30)
ax.set_xlabel('$p_T $ (GeV) ',fontsize=30) 
axes = plt.gca()
axes.set_ylim(-0.14,0.0)
plt.tight_layout()
savename='delta_sigma_mu_jec_norol_more_%s_%s%s'%(whats[which],sample,where)
plt.savefig(path+savename+'.pdf')
plt.savefig(path+savename+'.png')
plt.show()
```
