```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path='/scratch/snx3000/nchernya/bregression/plots/quantiles/2018-03-02/2018-03-01_job23_rawJetsJECtarget/ZHbbll/'
```

```python
whats=['p_T','eta','rho']
which = 0
```

```python
files = 'data_IQR_compare_%s_ZHbbll.csv'%whats[which]
```

```python
print(files)
```

```python
df = pd.read_csv(path+files)
for col in df.columns:
    if 'delta_' in col:
        df['%s_rolling'%col] = df[col].rolling(10).mean()
```

```python
df.round(4)
```

```python
ax = df.plot.scatter(x='%s'%whats[which],y='delta_ETH_Hbb_rolling',label='ETH variables vs Hbb variables')
df.plot.scatter(x='%s'%whats[which],y='delta_Hbb_JEC_rolling',color='r',ax=ax,label='Hbb vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_ETH_JEC_rolling',color='g',ax=ax,label='ETH vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_Pisa_JEC_rolling',color='b',ax=ax,label='Pisa vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_ETH_Pisa_JEC_rolling',color='c',ax=ax,label='ETH+Pisa vs JEC')

plt.show()
```

```python
ax = df.plot.scatter(x='%s'%whats[which],y='delta_sigma_ETH_Hbb_rolling',label='ETH variables vs Hbb variables')
df.plot.scatter(x='%s'%whats[which],y='delta_sigma_Hbb_JEC_rolling',color='r',ax=ax,label='Hbb vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_sigma_ETH_JEC_rolling',color='g',ax=ax,label='ETH vs JEC')
plt.show()
```

```python
ax = df.plot.scatter(x='%s'%whats[which],y='delta_Hbb_JEC_rel_rolling',color='r',label='Hbb vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_ETH_JEC_rel_rolling',color='g',ax=ax,label='ETH vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_Pisa_JEC_rel_rolling',color='b',ax=ax,label='Pisa vs JEC')
df.plot.scatter(x='%s'%whats[which],y='delta_ETH_Pisa_JEC_rel_rolling',color='c',ax=ax,label='ETH+Pisa vs JEC')

plt.show()
```
