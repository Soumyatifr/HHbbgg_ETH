```python
import numpy as np
import pandas as pd
```

```python
path = '/scratch/snx3000/nchernya/bregression/plots/dijet/'
#pT = [-40,40,60,70,80,90,100,110,120,150]
pT = [-40,40,60,80,85,90,95,100,105,110,120,150]
```

```python
lines = []
lines_jec = []
for pt in pT:
    if pt==-40: file = open(path+'mean_fwhm_.txt','r')
    else : file = open(path+'mean_fwhm_dijetPt>%d.txt'%pt,'r')
    idx = 0 
    for line in file:
        if idx==4 : lines.append(line)
        if idx==5 : lines_jec.append(line)
        idx+=1
```

```python
improvement = [float(lines[i][lines[i].find("-"):lines[i].find('\n')]) for i in range(0,len(lines))]
improvement_jec = [float(lines_jec[i][lines_jec[i].find("-"):lines_jec[i].find('\n')]) for i in range(0,len(lines_jec))]
print(improvement_jec)
print(improvement)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(pT,improvement,color='r',s=91,label='NN 2016 vs JEC')
plt.ylabel('Improvement in FWHM/peak wrt average ',fontsize=20)
plt.xlabel('cut on DiJet $p_T $ (GeV) ',fontsize=20) 
axes = plt.gca()
axes.set_ylim(-0.25,-0.10)
plt.tight_layout()
savename='improvement_DiJetcut'
plt.savefig(path+savename+'.pdf')
plt.savefig(path+savename+'.png')
plt.show()
```

```python
plt.scatter(pT,improvement_jec,color='r',s=91,label='NN 2016 vs JEC')
plt.ylabel('Improvement in FWHM/peak wrt JEC ',fontsize=20)
plt.xlabel('cut on DiJet $p_T $ (GeV) ',fontsize=20) 
axes = plt.gca()
axes.set_ylim(-0.25,-0.10)
plt.tight_layout()
savename='improvement_jec_DiJetcut'
plt.savefig(path+savename+'.pdf')
plt.savefig(path+savename+'.png')
plt.show()
```
