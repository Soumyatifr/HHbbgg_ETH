```python
inp = open(path+files[0],"r")
dictionaries = []
for i in range(0,len(files)):
    lines = []
    inp = open(path+files[i],"r")
    for line in inp:
        dicti = {}
        if '(+/-' in line :
            lines.append(line)
            dicti['value'] = float(line[:line.find(' (')])
            dicti['params'] = line[:line.find(' }')]
            dicti['fromFile#'] = i
            if (len(lines)<num[i]+1) : dictionaries.append(dicti) #because I printed twice the output 
    inp.close()
        
```
